```python
import pandas as pd
from transformers import T5Tokenizer,Trainer,T5ForConditionalGeneration,TrainingArguments
```

```python
train_data = pd.read_csv(r"/content/samsum-test.csv")
vali_data = pd.read_csv(r"/content/samsum-validation.csv")
train_data.head()
```

```python
train_data = train_data.sample(
    n=min(4000, len(train_data)),
    random_state=42
).reset_index(drop=True)

vali_data = vali_data.sample(
    n=min(500, len(vali_data)),
    random_state=42
).reset_index(drop=True)
```

```python
import re

def clean_data(text):
    text = re.sub(r"<.*?>", "", text)      # remove HTML tags
    text = re.sub(r"\r\n", " ", text)      # replace line breaks with space
    text = re.sub(r"\s+", " ", text)       # normalize spaces
    text = text.lower()                   # lowercase
    text = text.strip()                   # remove leading/trailing spaces
    return text
```

```python
train_data['dialogue']=train_data['dialogue'].apply(clean_data)
train_data['summary']=train_data['summary'].apply(clean_data)
vali_data['dialogue']=vali_data['dialogue'].apply(clean_data)
vali_data['summary']=vali_data['summary'].apply(clean_data)
```

```python
from transformers import T5Tokenizer

tokenizer = T5Tokenizer.from_pretrained("t5-small")
```

```python
def tokenize_function(data):
    input=tokenizer(data['dialogue'],padding="max_length",max_length= 512,truncation=True)
    target=tokenizer(data['summary'],padding='max_length',max_length=150,truncation=True)
    input['labels']=target["input_ids"]
    return input
```

```python
train_dataset=train_data.apply(tokenize_function,axis=1).tolist()
vali_dataset=vali_data.apply(tokenize_function,axis=1).tolist()
```

```python
model=T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
import torch
if torch.backends.mps.is_available():
    device=torch.device("mps")
elif torch.cuda.is_available():
    device=torch.device("cuda")
else:
    device=torch.device("cpu")
print(device)
model.to(device)
```

```python
train_arg=TrainingArguments(
    output_dir="./result",
    num_train_epochs=6,
    weight_decay=0.01,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="epoch",
    save_strategy="epoch",
    warmup_steps=500
)
trainer=Trainer(
    model=model,
    args=train_arg,
    train_dataset=train_dataset,
    eval_dataset=vali_dataset
)
```

```python
trainer.train()
```

```python
model.save_pretrained("./saved_memory_model")
tokenizer.save_pretrained("./saved_memory_model")
```
